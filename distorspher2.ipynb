```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

yv=np.linspace(0.1,1,20)
xv=np.linspace(-1,1,20)

x, y = np.meshgrid(xv, yv)
z = np.random.random(x.shape)

xdef, ydef = x/np.sqrt(x**2+y**2), y/np.sqrt(x**2+y**2)*y
xdef=(np.transpose(np.linspace(1,20,20)*np.transpose(xdef)))

#xdef, ydef = x*np.sqrt(1-y**2), y*np.sqrt(1-x**2)

fig, axes = plt.subplots(ncols=2)
axes[0].pcolormesh(x, y, z, cmap='gist_earth')
axes[1].pcolormesh(xdef, ydef, z, cmap='gist_earth')

axes[0].set(title='Original', xticks=[], yticks=[])
axes[1].set(title='Deformed', xticks=[], yticks=[])

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 13)
v = np.linspace(0, np.pi, 7)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, rstride=1, cstride=1, color='w', shade=0)

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

phiv=np.linspace(-np.pi/4,np.pi/4,24)
thev=phiv



phi, the=np.meshgrid(phiv,thev)
X=np.tan(phi)
Y=1/np.tan(the)*np.cos(phi)
delta=1+X**2+Y**2
C=np.sqrt(1+X**2)
D=np.sqrt(1+Y**2)

zrand = np.random.random(X.shape)
z=np.ones(X.shape)
#fig, axes = plt.subplots(ncols=2)
#%axes[0].pcolormesh(X, Y, z, cmap='gist_earth')
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Axsi=C*D/np.sqrt(delta)*phi
Aeta=-1*the + X*Y/np.sqrt(delta)*phi
ax.plot_surface(Axsi, Aeta, z, rstride=1, cstride=1, color='w', shade=0)

plt.show()
```
