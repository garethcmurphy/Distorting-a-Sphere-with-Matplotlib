```python
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm


xv=np.linspace(-1,1,20)
yv=xv
zv=xv

x, z = np.meshgrid(xv, zv)
y = -np.ones(z.shape)

xt=x* np.sqrt(  1 -0.5 *y**2 -0.5*z**2 + 0.3333* y**2 * z**2 ) 
yt=y* np.sqrt(  1 -0.5 *z**2 -0.5*x**2 + 0.3333* z**2 * x**2 ) 
zt=z* np.sqrt(  1 -0.5 *x**2 -0.5*y**2 + 0.3333* x**2 * y**2 ) 

###


y, z = np.meshgrid(yv, zv)
x = np.ones(y.shape)

xd=x* np.sqrt(  1 -0.5 *y**2 -0.5*z**2 + 0.3333* y**2 * z**2 ) 
yd=y* np.sqrt(  1 -0.5 *z**2 -0.5*x**2 + 0.3333* z**2 * x**2 ) 
zd=z* np.sqrt(  1 -0.5 *x**2 -0.5*y**2 + 0.3333* x**2 * y**2 ) 


xt=np.concatenate((xt, xd), axis=1)
yt=np.concatenate((yt, yd), axis=1)
zt=np.concatenate((zt, zd), axis=1)


###


y, z = np.meshgrid(yv, zv)
x = np.ones(z.shape)

xd=x* np.sqrt(  1 -0.5 *y**2 -0.5*z**2 + 0.3333* y**2 * z**2 ) 
yd=y* np.sqrt(  1 -0.5 *z**2 -0.5*x**2 + 0.3333* z**2 * x**2 ) 
zd=z* np.sqrt(  1 -0.5 *x**2 -0.5*y**2 + 0.3333* x**2 * y**2 ) 


xt=np.concatenate((xt, xd), axis=1)
yt=np.concatenate((yt, yd), axis=1)
zt=np.concatenate((zt, zd), axis=1)

####

x, z = np.meshgrid(-xv, zv)
y = np.ones(z.shape)

xd=x* np.sqrt(  1 -0.5 *y**2 -0.5*z**2 + 0.3333* y**2 * z**2 ) 
yd=y* np.sqrt(  1 -0.5 *z**2 -0.5*x**2 + 0.3333* z**2 * x**2 ) 
zd=z* np.sqrt(  1 -0.5 *x**2 -0.5*y**2 + 0.3333* x**2 * y**2 ) 


xt=np.concatenate((xt, xd), axis=1)
yt=np.concatenate((yt, yd), axis=1)
zt=np.concatenate((zt, zd), axis=1)


###


y, z = np.meshgrid(-yv, zv)
x = -np.ones(z.shape)

xd=x* np.sqrt(  1 -0.5 *y**2 -0.5*z**2 + 0.3333* y**2 * z**2 ) 
yd=y* np.sqrt(  1 -0.5 *z**2 -0.5*x**2 + 0.3333* z**2 * x**2 ) 
zd=z* np.sqrt(  1 -0.5 *x**2 -0.5*y**2 + 0.3333* x**2 * y**2 ) 


xt=np.concatenate((xt, xd), axis=1)
yt=np.concatenate((yt, yd), axis=1)
zt=np.concatenate((zt, zd), axis=1)

####

x1=np.ravel(xt)
y1=np.ravel(yt)
z1=np.ravel(zt)

fig = plt.figure(figsize=plt.figaspect(0.3))

ax = fig.add_subplot(1, 3, 1, projection='3d')
surf = ax.plot_surface(xt, yt, zt, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.plot_wireframe(xt, yt, zt, rstride=1, cstride=1)

ax = fig.add_subplot(1, 3, 3, projection='3d')
ax.plot_trisurf(x1, y1, z1)


```
