```python
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm


xv=np.linspace(-1,1,20)
yv=xv
zv=xv

x, y = np.meshgrid(xv, yv)
z = np.ones(x.shape)


xd=x* np.sqrt(  1 -0.5 *y**2 -0.5*z**2 + 0.3333* y**2 * z**2 ) 
yd=y* np.sqrt(  1 -0.5 *z**2 -0.5*x**2 + 0.3333* z**2 * x**2 ) 
zd=z* np.sqrt(  1 -0.5 *x**2 -0.5*y**2 + 0.3333* x**2 * y**2 ) 


fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax.plot_surface(xd, yd, zd, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

plt.plot_trisurf(xd,yd,zd)
```
